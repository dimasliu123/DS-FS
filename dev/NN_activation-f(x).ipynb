# Library :

```python
import tensorflow as tf
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.style.use('fivethirtyeight')
```

# Activation f(x) :

```python
X= tf.linspace(-10, 10, 100) # defining x (inputs)
```

## Activation for Hidden Layer :

### Identity :

```python
def identity(x):
    return x
```

```python
y= [identity(x) for x in X]
```

```python
plt.title('Identity')
plt.plot(X, y);
```

### Binary Step :
range(0,1)
- 0 = x < 0
- 1 = x >= 0

```python
def binary_step(x):
    if x < 0 :
        x = 0
    else :
        x = 1
    return x
```

```python
y = [binary_step(x) for x in X]
```

```python
plt.title('Binary Step')
plt.plot(X, y);
```

### ReLU :
- max(0.0, x)

```python
def reLU(x):
    return max(0.0, x)
y= [reLU(x) for x in X]
```

```python
plt.title('ReLU')
plt.plot(X, y);
```

### Leaky ReLU :

```python
def leaky_relu(x):
    if x < 0 :
        x = 0.01 * x
    elif x >= 0 :
        x = x
    return x
```

```python
y= [leaky_relu(x) for x in X]
```

```python
plt.title('Leaky ReLU')
plt.plot(X, y);
```

### Sigmoid :
range(0,1)
- 1.0 / (1.0 + e^-x)

```python
def sigmoid(x):
    return 1.0 / (1.0 + tf.math.exp(-x))
```

```python
y= [sigmoid(x) for x in X]
```

```python
plt.title('Sigmoid')
plt.plot(X, y)
plt.axhline(0.5, color='red', linestyle='dashed');
```

### Tanh :
- (e^x – e^-x) / (e^x + e^-x)

```python
def tanh(x):
    return ((tf.math.exp(x) - tf.math.exp(-x)) / (tf.math.exp(x) + tf.math.exp(-x)))
```

```python
y = [tanh(x) for x in X]
```

```python
plt.title('Hyperbolic Tangent (Tanh)')
plt.plot(X ,y);
```

### Sinusoid :
- sin(x)

```python
def sinusoid(x):
    return tf.math.sin(x)
```

```python
y = [sinusoid(x) for x in X]
```

```python
plt.title('Sinusoid', fontsize=16)
plt.plot(X, y);
```

### Gaussian :
range(0,1)
- e^-x**2

```python
def gaussian(x):
    return tf.math.exp(-x**2)
```

```python
y = [gaussian(x) for x in X]
```

```python
plt.title('Gaussian')
plt.plot(X, y);
```

### ArcTan :
range(-pi/2, pi/2)

- tan**-1(x)

```python
def arctan(x):
    return tf.math.atan(x)
```

```python
y= [arctan(x) for x in X]
```

```python
plt.title('ArcTan')
plt.plot(X, y);
```

### Soft Plus :
range(0, inf)
- log(1+ e^x)

```python
def softplus(x):
    return tf.math.log(1 +tf.math.exp(x))
```

```python
y = [softplus(x) for x in X]
```

```python
plt.title('Soft Plus')
plt.plot(X, y);
```

### Sigmoid Weighted Linear Unit (SILU) :
- x * (1 / 1 + e(-x))

```python
def silu(x):
    return x * (1.0 / 1.0 + tf.math.exp(-x))
```

```python
y = [silu(x) for x in X]
```

```python
plt.title('Sigmoid Weight Linear Unit (SILU)')
plt.plot(X, y);
```

### Parametric Rectified Linear Unit (PRELU) :

```python
def prelu(x, eta):
    if x < 0 :
        x = x * eta
    elif x >= 0 :
        x = x
    return x
```

```python
y = [prelu(x, 0.3) for x in X]
```

```python
plt.title('PRELU')
plt.plot(X, y);
```

### Bent Identity :

```python
def bent_identity(x):
    return tf.math.sqrt((x**2+1)-1)/2 + 1
```

```python
y = [bent_identity(x) for x in X]
```

```python
plt.title('Bent Identity')
plt.plot(X,y);
```

### Square Non Linearity (SQNL) :

```python
def sqnl(x):
    if x > 2.0 :
        x = 1
    elif 0 <= x <= 2.0 :
        x = x - (x**2/4)
    elif -2.0 <= x < 0 :
        x = x+ (x**2/4)
    else :
        x = -1
    return x
```

```python
y = [sqnl(x) for x in X]
```

```python
plt.plot(X, y);
```

### Inverse Square Root Unit (ISRU) :

```python
def inverse_square_root_unit(x, eta):
    return x / tf.math.sqrt(1+eta*(x**2))
```

```python
y = [inverse_square_root_unit(x, 0.3) for x in X]
```

```python
plt.title('Inverse Square Root Unit (ISRLU)')
plt.plot(X, y);
```

### Inverse Square Root Linear Unit (ISRLU) :

```python
def isrlu(x, eta):
    if x < 0 :
        x = x / tf.math.sqrt(1+eta*(x**2))
    elif x >= 0 :
        x= x
    return x
```

```python
y = [isrlu(x, 0.1) for x in X]
```

```python
plt.title('Inverse Square Root Linear Unit (ISRLU)')
plt.plot(X, y);
```

### Randomized Leaky Rectified Linear Unit (RRELU) :
x < 0 :
- a * x

x >= 0 :
- x

```python
def rrelu(x, eta) :
    if x < 0 :
        x = x * eta
    elif x >= 0 :
        x = x
    return x
```

```python
y = [rrelu(x, 0.3) for x in X]
```

```python
plt.title('RRELU')
plt.plot(X, y);
```

### Exponential Linear Unit (ELU):

```python
def elu(x, eta):
    if x <= 0 :
        x = x
    elif x > 0 :
        x = eta * (tf.math.exp(x) - 1)
    return x
```

```python
y = [elu(x, 0.01) for x in X]
```

```python
plt.title('Exponential Linear Unit')
plt.plot(X, y);
```

### Soft Clipping :

```python
def soft_clipping(x, eta):
    return 1/eta * (tf.math.log(1 + tf.math.exp(eta * x))) / (1 + tf.math.exp(eta * (x-1)))
```

```python
y = [soft_clipping(x, 0.001) for x in X]
```

```python
plt.title('Soft Clipping')
plt.plot(X, y);
```

### Soft Exponential :

```python
def soft_exponential(x, eta):
    if eta < 0 :
        x = -tf.math.log(1-eta*(x+eta)/eta)
    elif eta == 0 :
        x = x
    elif eta > 0 :
        x = ((tf.math.exp(eta*x) - 1)/ eta) + eta
    return x
```

```python
y = [soft_exponential(x, 0.3) for x in X]
```

```python
plt.title('Soft Exponential')
plt.plot(X,y);
```

### Soft Sign :

```python
def softsign(x):
    return x/(tf.math.abs(x)+1)
```

```python
y = [softsign(x) for x in X]
```

```python
plt.title('Soft Sign')
plt.plot(X, y);
```

### Sigmoid Output Activation Function :
- The same as above.

### Softmax Output Activation Function :
- e^x / sum(e^x)

```python
def softmax(x):
    return tf.math.exp(x) / tf.reduce_sum(tf.math.exp(x))
```

```python
y = softmax(X)
```

```python
plt.title('Softmax Output Activation Function')
plt.plot(X, y);
```
