```python
import numpy as np
from algo.preprocess import QuantizedScaling, Standardize, MinMax

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
plt.style.use("ggplot")
```

# Linear Data :

```python
X = np.linspace(0, 100, num=1000)
```

```python
mm_sc = MinMax()
mm_sc.calc(X)

q_sc = QuantizedScaling()
q_sc.calc(X)

z_sc = Standardize()
z_sc.calc(X)

X_q = q_sc.scale(X)
X_mm = mm_sc.scale(X)
X_zs = z_sc.scale(X)
```

```python
plt.figure(figsize=(14, 4.5))
plt.subplot(1, 2, 1)
plt.title('Linear Space Data')
plt.plot(X)

plt.subplot(1, 2, 2)
plt.title('Scaling Comparison on Linear Space')
sns.kdeplot(X_mm, shade=True, color='r', label='Min Max')
sns.kdeplot(X_q, shade=True, color='g', label='Quantized')
sns.kdeplot(X_zs, shade=True, color='b', label='Z Score')
plt.legend();
```

# Generating data with outlier :

```python
X2 = np.random.gamma(shape=5, scale=15, size=(100,))
```

```python
print("Mean :", X2.mean().round(2), "| Minimum :", X2.min().round(2), "| Median :", np.quantile(X2, 0.5).round(2), "| Maximum :", X2.max().round(2))
```

```python
plt.figure(figsize=(15, 5))
plt.subplot(1, 2, 1)
sns.boxplot(x=X2)
plt.subplot(1, 2, 2)
sns.kdeplot(x=X2, shade=True);
```

```python
mm_sc2 = MinMax()
mm_sc2.calc(X2)
q_sc2 = QuantizedScaling()
q_sc2.calc(X2)
z_sc2 = Standardize()
z_sc2.calc(X2)

X_q2 = q_sc2.scale(X2)
X_mm2 = mm_sc2.scale(X2)
X_zs2 = z_sc2.scale(X2)
```

```python
plt.figure(figsize=(14, 4.5))
plt.subplot(1, 2, 1)
plt.title('Gamma Distribution')
sns.kdeplot(X2, shade=True)

plt.subplot(1, 2, 2)
plt.title('Scaling Comparison on Gamma Distribution')
sns.kdeplot(X_mm2, shade=True, color='r', label='Min Max')
sns.kdeplot(X_q2, shade=True, color='g', label='Quantized')
sns.kdeplot(X_zs2, shade=True, color='b', label='Z Score')
plt.legend();
```
