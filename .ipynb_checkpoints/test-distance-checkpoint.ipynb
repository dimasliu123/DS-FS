```python
import numpy as np
from algo import distance
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
plt.style.use("ggplot")
```

# Generate random uniform distributed data :

```python
P = np.random.uniform(low=0, high=50, size=(100, 5)) # P -> 1st Vector Point
Q = np.random.uniform(low=0, high=50, size=(100, 5)) # Q -> 2nd Vector Point
```

```python
plt.figure(figsize=(8,5))
plt.title("Data Distribution")
for i in range(P.shape[1]):
    sns.kdeplot(P[:,i], fill=True, label=f"P{i}") # Label -> Vector variable : num_cols
    sns.kdeplot(Q[:,i], fill=True, label=f"Q{i}")
    plt.legend();
```

```python
euc_dist = distance.euclidean(P, Q) # Euclidean Distance 
man_dist = distance.manhattan(P, Q) # Manhattan Distance
min_dist = distance.minkowski(P, Q, 3) # Minkowski Distance
cheb_dist = distance.multi_dim_chebychev(P, Q) # 2D Chebychev distance
can_dist = distance.canberra(P, Q) # Canberra Distance
stand_euc_dist = distance.standard_euclidean(P, Q) # Standardized Euclidean Distance
chi2_dist = distance.chi_square_dist(P, Q) # Chi Squared Distance
```

```python
print("Distance Measurement : \n")
print("Euclidean Distance :", euc_dist)
print("Manhattan Distance :", man_dist)
print("Minkowski Distance :", min_dist)
print("Chebychev Distance :", cheb_dist)
print("Canberra Distance :", can_dist)
print("Standardized Euclidean Distance :", stand_euc_dist)
print("Chi2 Distance :", chi2_dist)
```
